```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Market_Basket_Optimisation.csv') 
# Python thinks the first row contains the headers

data = pd.read_csv('Market_Basket_Optimisation.csv', header=None) 
```

```python
data
```

```python
# It is needed to properly convert the data
transactions = []
for i in range(0,data.shape[0]):
    transactions.append([str(data.values[i,j]) for j in range(0,data.shape[1])])
    
```

```python
pd.DataFrame(transactions)
```

```python
# Import the apriori function
from apyori import apriori
```

```python
rules = apriori(transactions, 
                min_support = (3*7/7500),   #min support: num transactions/total transactions | one needs to observe and decide the frequency 
                min_confidence = 0.2,       #min confidence:if we set relatively higher, we get common-sense undesired/irrelevant rules
                min_lift = 3,                   #
                min_length = 2 ) 

  
```

```python
#Visualizing the Results

results = list(rules)
```
